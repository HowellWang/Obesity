```python
import pandas as pd
import matplotlib.pyplot as plt
%pylab inline
```

```python
tables = pd.read_html("https://www.cdc.gov/mmwr/volumes/65/wr/mm6507a1.htm")
```

```python
tables[0]
```

```python
df = tables[0]
```

```python
df.head(5)
```

```python
df1 = df.iloc[:, 5:8]
```

```python
s1 = df["State"]
```

```python
result = pd.concat([s1, df1], axis=1)
```

```python
old_names = list(result.columns)
new_names = ['State', 'None', '<1 time/day','≥1 time/day']
result.rename(columns=dict(zip(old_names, new_names)), inplace=True)
```

```python
result1 = result.copy()
```

```python
result1['None'] = result1['None'] .apply(lambda x : x[0:4])
result1['<1 time/day'] = result1['<1 time/day'] .apply(lambda x : x[0:4])
result1['≥1 time/day'] = result1['≥1 time/day'] .apply(lambda x : x[0:4])
```

```python
result1
```

```python
result1.to_csv("result1.csv")
```
